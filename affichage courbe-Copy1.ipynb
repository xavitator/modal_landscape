```python
from matplotlib import pyplot as plt
import numpy as np
```

# vgg_pretrained

```python
preci_test = np.load("courbe/preci_test_vgg_sup150.npy")
```

```python
preci_train = np.load("courbe/preci_train_vgg_sup150.npy")
```

```python
plt.plot(range(len(preci_train)), preci_train, label = "train")
plt.plot(range(len(preci_test)), preci_test, label = "test")
plt.legend()
plt.show()
```
